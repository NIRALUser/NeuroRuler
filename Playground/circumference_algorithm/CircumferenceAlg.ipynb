```python
!pip3 install opencv-python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img: np.ndarray = cv2.imread("sample.jpg")

# TODO:
# This need to be updated. 
# We may need several loops to find the best thresholds for different images.
edges: np.ndarray = cv2.Canny(img, 150, 500)
"""Binary bitmap with only contour pixels having value 255. Everything else has value 0."""

# with open('edges.txt', 'w') as f:
#     for i in range(len(edges)):
#         for j in range(len(edges[0])):
#             f.write(str(edges[i][j]))
#         f.write('\n')
 
cv2.imwrite('result.jpg', edges)

plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection Result')
```

```python
# Find contours.
contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
"""contours is a list of all the contours cv2 finds in the image. For example, contour[0] might represent an outer contour and contour[1] might represent an inner contour, if any.

The length of `contours` differs depending on the image."""

# print(f'type of contours[0]: {type(contours[0])}')
# with open('contours[0].txt', 'w') as f:
#     f.write(str(contours[0]))

# print(f'type of contours[1]: {type(contours[1])}')
# with open('contours[1].txt', 'w') as f:
#     f.write(str(contours[1]))

contour: np.ndarray = contours[0]
"""contours[0] is a numpy array that looks like [[0, 1], [1, 2], [3, 4]], representing the endpoints of coordinate points in the contour, I think. `cv2.arcLength` takes this and computes arc length from the endpoints."""
circumference = cv2.arcLength(contour, True)

contour2: np.ndarray = contours[1]
circumference2 = cv2.arcLength(contour2, True)

# How many detected edges?
# Keeping this under 2?
print(len(contours))
print('')

# Print the circumference.
# The difference of these two results can be used to evaluate whether the edge detection is successful or not.
print(f'Circumference: {circumference}')
print(f'Circumference: {circumference2}')
```
